#### Import Statements 

```python
import numpy as np
import pandas as pd
```

#### Loading the Data File 

```python
df = pd.read_csv("Matplotlib_SampleData/ETH_1h.csv")
```

```python
df ["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d %I-%p")
```

### Method 01 - Using Matplotlib from Scratch

```python
import matplotlib.pyplot as plt
from IPython.display import display, Image
import matplotlib
matplotlib.rcParams["text.usetex"] = True
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(df["Date"], df["Close"])
ax.set(ylabel="Close", xlabel="Date")
```

### Method 02 - Using Pandas integrated Plot Method | This is the way :)

```python
df.plot(x="Date", y="Open", kind="scatter", figsize=(10,5))
plt.text(s="ETH-1h", y=1400, x="2020-01", color="r")
```
