## Section 01 - Plotting Different Types of Graphs (The pyplot API Way) :

### Loading Data From Files

```python
x, y, p, q = np.loadtxt("Matplotlib_SampleData/sample_data.txt", delimiter=",", unpack=True)
```

```python
df = pd.read_csv("Matplotlib_SampleData/survey_results_public.csv")
```

### 1) Introduction (line graph, legends, titles and labels)

```python
plt.plot(x, y, label="First Graph")  # Simple Line plot
plt.plot(p, q, label="Second Graph")  # Label is used so that .legend can identify the graphs separately
plt.legend()
plt.title("Introductory Graph\nCheck it Out")
plt.xlabel("Independent Var")
plt.ylabel("Dependent Var")
```

### 2) Bar Charts & Histograms

```python
bar_samplex= ["Maidul", "Moinul", "Palas", "Pial"]
bar_sampley= [80, 90, 95, 80]
```

```python
plt.bar(bar_samplex, bar_sampley, label="room-204", color="c")
```

```python
hist_dist = [2, 5, 52, 65, 18, 30, 6, 2, 45, 32, 90, 18, 29, 21, 23, 14, 42, 32, 17, 21, 22]
hist_bins = [x*10 for x in range(0, 11)]
plt.hist(hist_dist, hist_bins, histtype="bar", rwidth=.8, orientation="vertical")
```

### 3) Scatter Plots

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [4, 7, 8, 3, 5, 3, 5, 1, 8]
plt.scatter(x, y, s=50, marker="X")  # s is the param for defining marker size
```

### 4) Stack Plots

```python
days = ["sat", "sun", "mon", "tue", "wed", "thu", "fri"]
sleeping = [8, 7, 6, 6, 7, 9, 10]
eating = [2, 3, 2, 2, 3, 3, 4]
working = [2, 3, 2, 3, 3, 5, 7]
others = [12, 11, 14, 13, 11, 7, 3]
```

```python
plt.plot([], [], color="m", label="sleeping", linewidth=5)
plt.plot([], [], color="c", label="eating", linewidth=5)
plt.plot([], [], color="r", label="working", linewidth=5)
plt.plot([], [], color="k", label="others", linewidth=5)
plt.legend()

plt.stackplot(days, sleeping,eating,working,others, colors=["m", "c", "r", "k"])
plt.xlabel("Day")
plt.ylabel("Time Spent (in Hours)")
plt.title("A Stack Plot descriving the time spent portion of a Day in the\n3rd week of JUN 2020")
```

### 5) Pie Charts

```python
race = ["Hispanic", "White", "African American", "American Indian", "Asian/Pacific", "Other"]
num_students = [40, 35, 12, 15, 7, 11]
plt.pie(num_students, labels=race, 
        autopct="%1.2f%%", 
        radius=1.5, 
        explode=(.1, .1, .1, .1, .1, .1), 
        shadow=True,
        colors=["r", "g", "b", "c", "m", "y"]
       )
plt.show()
```

### 1) Plotting Different Types of Graphs 

```python
plt.style.use("seaborn-whitegrid")

# Creating & Customizing the Figure 
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(20, 10))
fig.suptitle("Differeent Types of Useful Plots", fontsize=16, fontweight="bold")

# Line Plot (plotting on 1st axes of the figure)
ax0.set(xlabel="Independent Var", ylabel="Dependent Var", title="Line Plot")
ax0.plot(x,y, color="c")
ax0.legend("plot.legend_elements()")

# Bar Chart (plotting on 2nd axes of the figure)
ax1.set(xlabel="Name", ylabel="Score", title="Bar Chart")
ax1.bar(bar_samplex, bar_sampley, label="room-204", color="g", width=.5)
ax1.legend("bar.legend_elements()")

# Histogram (plotting on 3rd axes of the figure)
ax2.set(xlabel="Frequency", ylabel="Bins", title="Histograms")
ax2.hist(hist_dist, hist_bins, histtype="bar", rwidth=.8, orientation="horizontal")
ax2.legend("hist.legend_elements()")

# Scatter Plot (plotting on 4th axes of the figure)
ax3.set(xlabel="X", ylabel="Y", title="Scatter Plot")
ax3.scatter(x, y, s=50, marker="o", color="r")
ax3.hlines(y=[min(x), max(x), np.mean(x)],  xmin=0, xmax=10,
           colors=["c", "c", "b"], linestyles={"--"})
ax3.legend("scatter.legend_elements()")
```
