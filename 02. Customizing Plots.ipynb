```python
x = np.linspace(0, 2, 100)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
fig.suptitle(r"\underline{Demo Plot}", fontsize="20", fontweight="bold", fontstyle="italic")

# plot in subplot ax
ax.plot(x, x**3, label="blue", color="b")
ax.plot(x, x**2, label="green", color="g")
ax.plot(x, x, label="cian", color="c")

ax.text(1.5, 2, r"$X^2$")
ax.annotate(r"$$X^3$$", (2, 8), xytext=(1.5, 8),arrowprops=dict(facecolor="grey", width=1, color="grey"))

ax.set_xlabel(r'$\rho/\rho_{ref}\;\rightarrow$', color='red', fontsize=16, fontweight="bold")
ax.set_ylabel(r'$\Delta \Theta / \omega \longrightarrow$', color='red', fontsize=16, fontweight="bold")
ax.set_title("Subplot ax")
plt.sca(ax)
plt.xticks(ticks=[0., .5, 1., 1.5, 2.], labels=[0, "a", "b", "c", "d"] ,rotation=45, color="b")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

## Section 02 - Fill between Plots (Customization of fill): 

```python
df = pd.read_csv("Matplotlib_SampleData/ETH_1h.csv")
```

```python
df ["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d %I-%p")
```

#### (?) Example : 
Let's say we bought some share of a certain crypto-currency at $200 per unit and we would like to know at which Dates did we gain Profit or Lost money from our vanture. Graphically represent the data to serve our needs.

```python
df.plot("Date", "Close", kind="line", color="g", figsize=(20, 10))
plt.axhline(200, color="k", linestyle="--")
plt.fill_between(df["Date"], df["Close"], 200,
                 where=(df["Close"] > 200), 
                 facecolor="c",
                 alpha=0.5)
plt.fill_between(df["Date"], df["Close"], 200,
                 where=(df["Close"] < 200), 
                 facecolor="r",
                 alpha=0.5)
plt.plot([], [], color="c", label="Profit")
plt.plot([], [], color="r", label="Loss")
plt.legend();
```

## Section 03 - Customize Date Plots:
